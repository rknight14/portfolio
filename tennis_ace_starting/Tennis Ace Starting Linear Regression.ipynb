# Tennis Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Load and investigate the data here:

```python
tennis = pd.read_csv('tennis_stats.csv')
tennis.head()
```

Perform exploratory analysis here:

```python
plt.scatter(tennis[['Winnings']],tennis[['BreakPointsOpportunities']])
```

Perform single feature linear regressions here:

```python
feature = tennis[['BreakPointsOpportunities']]
outcome = tennis[['Winnings']]
feature_train, feature_test, outcome_train, outcome_test = train_test_split(feature, outcome, train_size=0.8, test_size=0.2)

model = LinearRegression()
model.fit(feature_train, outcome_train)
model.score(feature_test,outcome_test)
outcome_predicted = model.predict(feature_test)
plt.scatter(outcome_test, outcome_predicted, alpha=0.4)
```

Perform two feature linear regressions here:

```python
feature2 = tennis[['FirstServeReturnPointsWon']]
outcome2 = tennis[['Winnings']]
feature_train2, feature_test2, outcome_train2, outcome_test2 = train_test_split(feature2, outcome2, train_size=0.8, test_size=0.2)

model2 = LinearRegression()
model2.fit(feature_train2, outcome_train2)
model2.score(feature_test2,outcome_test2)
outcome_predicted2 = model2.predict(feature_test2)
plt.scatter(outcome_test2, outcome_predicted2, alpha=0.4)
```

Perform multiple feature linear regressions here:

```python
feature3 = tennis[['BreakPointsOpportunities', 'FirstServeReturnPointsWon']]
outcome3 = tennis[['Winnings']]
feature_train3, feature_test3, outcome_train3, outcome_test3 = train_test_split(feature3, outcome3, train_size=0.8, test_size=0.2)

model3 = LinearRegression()
model3.fit(feature_train3, outcome_train3)
model3.score(feature_test3,outcome_test3)
outcome_predicted3 = model3.predict(feature_test3)
plt.scatter(outcome_test3, outcome_predicted3, alpha=0.4)
```
